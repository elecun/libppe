```python
import matplotlib.pyplot as plt

X = [160, 150, 140, 130, 120, 110, 100, 90, 80, 70, 60, 50]
y = [0, 30, 60, 90, 120, 180, 230, 280, 380, 480, 620, 800]

plt.plot(X,y,"b.")
plt.title("Relationship between FM Position and W/D", fontsize=12)
plt.ylabel("Focus Module Position(um)", fontsize=11)
plt.xlabel("Working Distance(mm)", fontsize=11)
plt.show()
```

```python
# Polynomial regression
import numpy as np

model = np.poly1d(np.polyfit(X, y, 2)) # 2-order
line = np.linspace(50, 160, len(X))

plt.scatter(X, y)
plt.plot(line, model(line))
plt.show()

print(model(line))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

X = np.arange(160, 40, -10)
y = np.array([1, 30, 60, 90, 120, 180, 230, 280, 380, 480, 620, 800])

model = np.polyfit(X, np.log(y), 1, w=np.sqrt(y))
a = np.exp(model[1])
b = model[0]
x_fitted = np.linspace(np.min(X), np.max(X), 12)
y_fitted = a * np.exp(b * x_fitted)

print("a:", a)
print("b:", b)

ax = plt.axes()
ax.scatter(X, y, label='Data')
ax.plot(x_fitted, y_fitted, 'k--', label='Exp. Reg Curve')
ax.set_title('Exponential Regression Results')
ax.set_ylabel('Focus Module Position(um)')
ax.set_ylim(0, 900)
ax.set_xlabel('Working Distance(mm)')
ax.legend()

print("R-Squared : ", r2_score(y, np.flip(y_fitted)))
```
