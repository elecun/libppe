```python
'''
Image Augmentation (Blur, Rotation, Scale) 
'''
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

tip = cv2.imread("../../data/forktip/forktip.png")
cv2.cvtColor(tip, cv2.COLOR_RGB2GRAY, tip)

plt.imshow(tip)
plt.show()
```

```python
''' 
Generate rotated images
'''
import imutils

filelist = []
for i in range(360):
    tip_rotated = imutils.rotate(tip, i)
    filename = "../../data/forktip/forktip_r_"+str(i)+".png"
    filelist.append(filename)
    cv2.imwrite(filename, tip_rotated)
```

```python
''' 
Generate blurred images
'''
import os

kernel_size = 51
step = 10
list_step = np.arange(1, step+1, 1, dtype=int)
for file in filelist:
    print("generating for ", file)
    tip_sample = cv2.imread(file)
    _path, _ext = os.path.splitext(file)
    for s in list_step:
        tip_blurred = cv2.GaussianBlur(tip_sample, (kernel_size, kernel_size), step)
        filename = _path+"_k"+str(kernel_size)+"_s"+str(s)+".png"
        cv2.imwrite(filename, tip_blurred)
```
