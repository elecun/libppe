```python
# format conversion from json mini to r-cnn annotation
# onlu single bounding box & single key point

import json
import os
import argparse
import glob


_annotation_path = "./train/annotations"
_input_filepath = os.path.join("./train", "annotations.json") # file from label studio json mini

#_image_files = sorted(glob.glob(os.path.join(_root_path, "images", "*.png")))
#_annotation_files = sorted(glob.glob(os.path.join(_root_path, "annotations", "*.txt")))

# format
# {"bboxes": [[962, 350, 1462, 534], [374, 490, 922, 734]], "keypoints": [[[1019, 487, 1], [1432, 404, 1]], [[861, 534, 1], [392, 666, 1]]]}

if os.path.isfile(_input_filepath):
    with open(_input_filepath, 'r') as f:
        json_data = json.load(f)

        for data in json_data:
            out_format = {}
            output_anno_file = os.path.splitext(os.path.join(_annotation_path, data['img']))[0]+".json"
#            print("created ", output_anno_file)

            x = data["label"][0]["x"]
            y = data["label"][0]["y"]
            w = data["label"][0]["width"]
            h = data["label"][0]["height"]
            org_width = data["label"][0]["original_width"]
            org_height = data["label"][0]["original_height"]

            x_min = x*org_width/100
            y_min = y*org_height/100
            x_max = x_min + w*org_width/100
            y_max = y_min + h*org_height/100
            
            out_format["bboxes"] = [[int(x_min), int(y_min), int(x_max), int(y_max)]]

            kp_x = data["kp-1"][0]["x"]*org_width/100
            kp_y = data["kp-1"][0]["y"]*org_height/100
            out_format["keypoints"] = [[[int(kp_x), int(kp_y), 1]]]
            
            with open(output_anno_file, 'w') as fo:
                out_format_str = json.dumps(out_format)
                fo.writelines(out_format_str)
            
            print("created annotation file : ", output_anno_file)

            
```
